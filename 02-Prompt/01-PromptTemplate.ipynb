```python
load_dotenv()
```

```python
logging.langsmith("langchain-02")  # 프로젝트 이름
```

```python
llm = ChatOpenAI()
```

1. from_template() 메소드로 PromptTemplate 객체 생성
- 치환될 변수를 {변수}로 묶어서 템플릿 정의

```python
template1 = "{country}의 수도는 어디인가요?"  # {country}가 변수
prompt1 = PromptTemplate.from_template(template1)
```

```python
prompt1.format(country="대한민국")
```

```python
template1 = "{country}의 수도는 어디인가요?"  # {country}가 변수
prompt1 = PromptTemplate.from_template(template1)

chain1 = prompt1 | llm
```

```python
chain1.invoke("대한민국").content  # 변수에 값 넣어주기
```

2. PromptTemplate 객체 생성과 동시에 prompt 생성

```python
template1 = "{country}의 수도는 어디인가요?"

prompt1 = PromptTemplate(
    template=template1, 
    input_variables=["country"]  # 추가 유효성 검사 목적. 인스턴스화 중에 템플릿 문자여렝 있는 변수와 비교해서 불일치 하는 경우 예외를 발생.
)
```

```python
prompt1
```

```python
prompt1.format(country="대한민국")  # country 변수에 값 미리 체워넣기
```

```python
template2 = "{country1}과 {country2}의 수도는 각각 어디인가요?"
prompt2 = PromptTemplate(
    template=template2,
    input_variables=["country1"],
    partial_variables={"country2": "미국"}  # country2를 '미국'으로 확정지어버리기
)
```

```python
prompt2
```

```python
prompt2.format(country1="대한민국")
```

```python
prompt_partial = prompt2.partial(country2="캐나다")
```

```python
prompt_partial
```

```python
prompt_partial.format(country1="대한민국")
```

```python
chain = prompt_partial | llm
```

```python
chain.invoke("대한민국").content
```

```python
chain.invoke({"country1": "대한민국", "country2": "호주"}).content
```
