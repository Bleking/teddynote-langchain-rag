```python
from dotenv import load_dotenv

from langchain_teddynote import logging
from langchain_openai import ChatOpenAI
```

```python
load_dotenv()
```

```python
# 프로젝트 이름을 입력합니다.
logging.langsmith("langchain-1.2")
```

```python
llm_with_logprob = ChatOpenAI(
    temperature=0.1, 
    max_tokens=2048, 
    model_name="gpt-4.1-nano", 
).bind(logprobs=True)
```

```python
# 질의내용
question = "대한민국의 수도는 어디인가요?"

# 질의
response_logprob = llm_with_logprob.invoke(question)
```
