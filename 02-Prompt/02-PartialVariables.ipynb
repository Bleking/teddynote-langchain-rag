partial_variables
- 공통된 방식으로 가져오고 싶은 변수가 있는 경우, 함수를 부분적으로 사용하기 위함
- 예를 들어, 항상 현재 날짜를 반환하는 함수를 사용해서 프롬프트를 하드코딩하지 않고 항상 현재 날짜를 표시할 수 있음

```python
from dotenv import load_dotenv

from langchain_teddynote import logging
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

from datetime import datetime
```

```python
load_dotenv()
```

```python
logging.langsmith("langchain-02")  # 프로젝트 이름
```

```python
llm = ChatOpenAI()
```

```python
def get_today():
    return datetime.now().strftime("%B %d")  # 오늘 날짜
```

```python
prompt = PromptTemplate(
    template="오늘의 날짜는 {today} 입니다. 오늘이 생일인 유명인 {n}명을 나열해 주세요. 생년월일을 표기해주세요.", 
    input_variables=["n"], 
    partial_variables={
        "today": get_today
    }
)
```

```python
prompt.format(n=3)
```

```python
chain = prompt | llm
```

```python
chain.invoke(3).content
```

```python
chain.invoke({"today": "Jan 02", "n": 3}).content  # 생일이 1월 2일이 유명일 3명
```
